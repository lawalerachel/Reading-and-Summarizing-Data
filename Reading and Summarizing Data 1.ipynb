## Reading and Summarizing Data

```python
import pandas as pd
order=pd.read_excel("orders 1.xlsx")
order
```

```python
# Bring out the data types of each column
order.dtypes
```

```python
# count the number of times order takes place in each destination
# value_counts ()
order["Destination"].value_counts()
```

```python
# Find out the total sum of numbers of items
order["Number of items"].sum(axis=0)
```

```python
# Count the number of different items
order['Item'].value_counts()
```

```python
# sort the driver's name in ascending order
# sort_values
order_sorted=order.sort_values("Driver's name").reset_index(drop=True)
order_sorted
```

```python
# Extract rows where the driver's name is 'John may'
#loc
order_sorted.loc[order_sorted["Driver's name"]=="John May"]
```

```python
# Extract rows where the items is washing machine
order_sorted.loc[order_sorted["Item"]=="washing machine"]
```

```python
# Extract rows where the number of item is 25 and more
order_sorted.loc[order_sorted['Number of items']>=25]
```
